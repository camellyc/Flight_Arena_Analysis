```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import sem
from matplotlib.ticker import MultipleLocator

sns.set_style("ticks")
sns.despine()

file_path = '/Users/camellyc/Downloads/240101_web_WtoR.xlsx'
df = pd.read_excel(file_path)

x = df['x_time']
y1 = df['web_signal']
y2 = df['hook_signal']
y1_sem = df['web_sem']
y2_sem = df['hook_sem']
```

```python
# Assuming y1_sem, y2_sem, y3_sem are the SEM arrays for y1, y2, y3 respectively
# You should replace these with your actual SEM calculations
linewidth_plot = 5
linewidth = 2
plt.style.use('dark_background') 
plt.figure(figsize=(8, 6), dpi=300)
x_min, x_max = np.min(x), np.max(x)
plt.xlim(x_min, x_max)
plt.ylim(0.15, 0.7)
plt.fill_betweenx(y=[0, 0.8], x1=np.max(x), x2=0, color='#7F7F7F', alpha=1)

plt.plot(x, y1, color='#76D6FF',lw=linewidth_plot)
plt.fill_between(x, y1 - y1_sem, y1 + y1_sem, color='#76D6FF', alpha=0.2)

plt.plot(x, y2, color='#009999',  lw=linewidth_plot)
plt.fill_between(x, y2 - y2_sem, y2 + y2_sem, color='#009999', alpha=0.2)


plt.legend(frameon=False, loc = 'lower right')

ax = plt.gca()
ax.spines['top'].set_linewidth(linewidth)
ax.spines['bottom'].set_linewidth(linewidth)
ax.spines['left'].set_linewidth(linewidth)
ax.spines['right'].set_linewidth(linewidth)
ax.xaxis.set_tick_params(width=linewidth, labelsize=20)
ax.yaxis.set_tick_params(width=linewidth, labelsize=20)

ax.yaxis.set_major_locator(MultipleLocator(base=0.1)) 
ax.xaxis.set_major_locator(MultipleLocator(base=0.25))  

sns.despine(offset=20, trim=False)
plt.tight_layout()
plt.show()
```
