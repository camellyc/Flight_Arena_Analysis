```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
file_path = '/Users/camellyc/Downloads/Web, Oval, BDN2 hemilineage MANC.xlsx'  # Update with the correct file path
df = pd.read_excel(file_path)

color = '#76D6FF'  # You can use color names or hexadecimal color codes
alpha_web = 1  # Adjust the alpha value for 'web' bars
alpha_oval = 1  # Adjust the alpha value for 'oval' bars
alpha_BDN2 = 1  # Adjust the alpha value for 'bdn2' bars

# Create a figure with three subplots, sharing the x-axis
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 9), dpi=300, sharex=True)

# Plotting the 'web' column
axes[0].bar(df['Hemilineage'], df['Web'], color='#7030A0', alpha=alpha_web)
axes[0].set_ylim(0, 4200)
axes[0].set_ylabel('Web Synapse Count')
axes[0].set_title('Number of Synapses with Different Hemilineage')

# Plotting the 'oval' column
axes[1].bar(df['Hemilineage'], df['Oval'], color='#CC922E', alpha=alpha_oval)
axes[1].set_ylim(0, 4200)
axes[1].set_ylabel('Oval Synapse Count')

# Plotting the 'bdn2' column
axes[2].bar(df['Hemilineage'], df['BDN2'], color='#4983C5', alpha=alpha_BDN2)
axes[2].set_ylim(0, 4200)
axes[2].set_ylabel('BDN2 Synapse Count')

# Set the x-axis label and rotate the labels only on the last subplot
axes[2].set_xlabel('Hemilineage')
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```
