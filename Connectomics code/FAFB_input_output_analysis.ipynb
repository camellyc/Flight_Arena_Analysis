```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

NOI = [720575940623786376,720575940645176264,720575940605702944,720575940624801771,720575940623142426,720575940637618142,720575940644666660,720575940619757366,720575940628659049,720575940622933765]

classification = pd.read_csv('/Users/camellyc/Desktop/Connectomics/FAFB/classification.csv', index_col=0, header = 0)
neurons = pd.read_csv('/Users/camellyc/Desktop/Connectomics/FAFB/neurons.csv', index_col=0, header = 0)
connections = pd.read_csv('/Users/camellyc/Desktop/Connectomics/FAFB/connections.csv', index_col=0, header = 0)
```

```python
# Filter the connections to include only those where NOI is the postsynaptic neuron
filtered_connections = connections[connections['post_root_id'].isin(NOI)]

# Create a pivot table to represent the connectivity matrix
connectivity_matrix = filtered_connections.pivot_table(index='pre_root_id', columns='post_root_id', values='syn_count', aggfunc='sum', fill_value=0)

# Plotting the connectivity matrix
plt.figure(figsize=(10, 8))
sns.heatmap(connectivity_matrix, cmap='viridis', annot=True, fmt='g')
plt.title('Connectivity Matrix with NOI as Postsynaptic Neurons')
plt.xlabel('Postsynaptic Neurons (NOI)')
plt.ylabel('Presynaptic Neurons')
plt.show()
```
