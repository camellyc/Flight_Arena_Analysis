```python
# Function to process and plot a summary figure
def process_and_plot_summary(folder_path, sampling_rate, target_duration, pre_stim_time, post_stim_time):
    # Load the CSV files from the folder
    all_data = load_csv_files(folder_path)

    # Initialize lists to store data across all trials
    all_wbf_windows = []
    all_spike_rate_windows = []
    all_stim_windows = []
    all_spike_count_windows = []
    time_windows = None  # To store aligned time windows for the first time

    for data in all_data:
        file, time, wingbeat_freq_smoothed, transformed_opto_stim, spike_rate, spike_count = data

        # Extract stimulus blocks where the duration matches the target duration
        stim_blocks = extract_stim_blocks(transformed_opto_stim, sampling_rate, target_duration)

        # Only proceed if there are blocks detected
        if stim_blocks:
            for block_start, block_end in stim_blocks:
                stim_on_time = time[block_start]
                start_idx = max(0, block_start - int(pre_stim_time * sampling_rate))
                end_idx = min(len(time), block_start + int(post_stim_time * sampling_rate))

                # Extract windowed data for each variable
                time_window = time[start_idx:end_idx] - stim_on_time  # Align to stim onset
                wingbeat_freq_window = wingbeat_freq_smoothed[start_idx:end_idx]
                spike_rate_window = spike_rate[start_idx:end_idx]
                transformed_stim_window = transformed_opto_stim[start_idx:end_idx]
                spike_count_window = spike_count[start_idx:end_idx]

                # Append extracted data to the lists
                all_wbf_windows.append(wingbeat_freq_window)
                all_spike_rate_windows.append(spike_rate_window)
                all_stim_windows.append(transformed_stim_window)
                all_spike_count_windows.append(spike_count_window)

                # Store time_window once (it will be the same for all blocks)
                if time_windows is None:
                    time_windows = time_window

    # Calculate mean and SEM for Wingbeat Frequency, Spike Rate, and Stimulus
    mean_wbf, sem_wbf = calculate_mean_and_sem(all_wbf_windows)
    mean_spike_rate, sem_spike_rate = calculate_mean_and_sem(all_spike_rate_windows)
    mean_stim, sem_stim = calculate_mean_and_sem(all_stim_windows)
    n_trials = len(all_wbf_windows)  # Number of trials used in the summary

    # Create a summary plot
    fig, ax1 = plt.subplots(figsize=(15, 5))

    # Plot Wingbeat Frequency with SEM
    plot_mean_with_sem(time_windows, mean_wbf, sem_wbf, color='blue', label='Wingbeat Frequency', ylabel='Wingbeat Frequency (Hz)', n_blocks=n_trials)

    # Plot Spike Rate with SEM on a second y-axis
    ax2 = ax1.twinx()
    plot_mean_with_sem(time_windows, mean_spike_rate, sem_spike_rate, color='green', label='Spike Rate', ylabel='Spike Rate (spikes/sec)', n_blocks=n_trials)
    ax2.set_ylim(0, 15)

    # Plot Optogenetic Stimulus on the primary axis with a fill
    ax1.fill_between(time_windows, 0, 250 * mean_stim, color='crimson', alpha=0.1, label=f'Optogenetic Stimulus (n={n_trials})')
    
    # Final plot adjustments
    ax1.set_xlabel('Time (seconds)')
    ax1.set_ylim(100, 260)
    fig.legend(loc='upper right', bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)
    plt.title('WBF and DLMn Spike Rate')
    plt.tight_layout()
    plt.show()

    ### Raster Plot with Shaded Stimulus Region ###
    # Create a shorter raster plot by setting a fixed height for each trial
    fig, ax = plt.subplots(figsize=(15, 5))  # Set a fixed height for the raster plot

    # Plot spikes for each trial
    for idx, spike_count_window in enumerate(all_spike_count_windows):
        spike_times = np.where(spike_count_window == 1)[0]  # Spike indices in the window
        spike_times_aligned = time_windows[spike_times.astype(int)]  # Align spike times to the current window
        ax.eventplot(spike_times_aligned, lineoffsets=idx, linelengths=0.9, color='black')  # Plot each row
    
    # Add shading for the stimulus region where opto-stim is on
    stim_on_indices = np.where(mean_stim > 0)[0]
    if len(stim_on_indices) > 0:
        stim_start = time_windows[stim_on_indices[0]]
        stim_end = time_windows[stim_on_indices[-1]]
        ax.axvspan(stim_start, stim_end, color='crimson', alpha=0.2, label='Stimulus On')  # Shaded region

    ax.set_xlabel('Time (seconds)')
    ax.set_ylabel('Trial')
    ax.set_xlim(time_windows[0], time_windows[-1])  # Ensure x-axis matches the summary plot
    ax.legend(loc='upper right')
    plt.title('DLMn Spike Raster')
    plt.tight_layout()
    plt.show()

process_and_plot_summary(folder_path, sampling_rate, target_duration, pre_stim_time, post_stim_time)
```
